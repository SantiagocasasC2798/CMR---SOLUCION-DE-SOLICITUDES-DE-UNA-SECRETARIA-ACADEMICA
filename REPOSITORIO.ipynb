```python
# @title Tipo de solicitud
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)

counts = [4,16,36,44,]
labels = ['Financiera','Admisiones','Semestre academico','Requisito para obtener el título']
pl.pie(counts, labels=labels, autopct=lambda pct: func(pct, counts))
pl.show()
```

```python
# @title Respuestas solicitudes realizadas en el ultimo año

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Hoja de cálculo sin título - Hoja 1 (1).csv')
data
```

```python
data['CUANTITATIVO'].mean()
```

```python
data['CUANTITATIVO'].std()
```

```python
# @title HISTOGRAMA

fig = pl.figure(figsize=(4,4))
pl.hist(data['CUANTITATIVO'], color='green', histtype='step')
pl.axvline(np.quantile(data.CUANTITATIVO, .40), c='black', label=f'P40: {np.quantile(data.CUANTITATIVO, 0.40)}') #Q1
pl.axvline(np.quantile(data.CUANTITATIVO, .60), c='blue', label=f'P60: {np.quantile(data.CUANTITATIVO, 0.60)}') #Q2
pl.legend()
pl.xlabel('Conformidad de las solicitudes')
```

```python
Q1 = np.quantile(data['CUANTITATIVO'], 0.40)
Q2 = np.quantile(data['CUANTITATIVO'], 0.60)

Q1, Q2
```

```python
IQR = Q2 - Q1
IQR
```

```python
ii = data.CUANTITATIVO < Q1 - 1.5*IQR
jj = data.CUANTITATIVO < Q2 + 1.5*IQR

sum(ii), sum(jj)
```

```python
Q1 = np.quantile(data.CUANTITATIVO, 0.40)
Q1 = np.quantile(data.CUANTITATIVO, 0.60)
IQR = Q2 - Q1

ii = data.CUANTITATIVO < Q1 - 1.5*IQR
jj = data.CUANTITATIVO < Q2 + 1.5*IQR

sum(ii), sum(jj)

data.CUANTITATIVO[ii | jj]
```

```python
_ = pl.violinplot(data.CUANTITATIVO, quantiles=[0.40, 0.60])
```
